```python
#importing the necessary libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#loading the data
data = pd.read_csv("data.csv")
data.head(10)
```

```python
#inspecting the data
data.info()
```

```python
data.describe()
```

```python
#Q1. Which countries have the highest suicide rates per 100k population?

top_rates = data.groupby("Country")["Suicides number"].mean()
top_rates = top_rates.sort_values(ascending= False)
print(top_rates)
```

```python
# Q2. Global trend over time

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Chart 1 — Raw numbers
sns.lineplot(data=data, 
             x='Year', 
             y='Suicides number',
             errorbar=None,
             color='darkred',
             linewidth=2.5,
             ax=axes[0])
axes[0].set_title('Total Suicide Numbers Over Time',
                  fontweight='bold')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Total Suicides')
axes[0].grid(True, alpha=0.3)

# Chart 2 — Per 100k
yearly = data.groupby('Year')['Suicides number'].mean().reset_index()
sns.lineplot(data=yearly,
             x='Year',
             y='Suicides number', 
             errorbar=None,
             color='darkblue',
             linewidth=2.5,
             ax=axes[1])
axes[1].set_title('Suicide Rate per 100k Population Over Time',
                  fontweight='bold')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Suicides per 100k')
axes[1].grid(True, alpha=0.3)

plt.suptitle('Global Suicide Trends — Two Ways of Looking at It',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Q3. Does alcohol consumption correlate with suicide rates?

plt.figure(figsize=(7, 5))
sns.scatterplot(data=data, 
                x='Alcohol', 
                y='Suicides number',
                alpha=0.4,
                color='purple')

# Adding trend line
sns.regplot(data=data,
            x='Alcohol',
            y='Suicides number',
            scatter=False, 
            color='red',
            line_kws={'linewidth': 2, 'linestyle': '--'})

plt.title('Countries With Higher Alcohol Consumption\nHave Higher Suicide Rates',
          fontsize=14,
          fontweight='bold',
          pad=15)
plt.xlabel('Alcohol Consumption', fontsize=12)
plt.ylabel('Suicides per 100k Population', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Printing the correlation value
corr = data['Alcohol'].corr(data['Suicides number'])
print(f"Correlation between Alcohol and Suicide Rate: {corr:.2f}")
```
